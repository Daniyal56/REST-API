```python
import csv
import json
```

```python
with open('kps.csv','r') as f:
    reader = csv.reader(f)
    data = {'items':[]}
    for row in reader:
        data['items'].append({'mchdb_pk':row[0],'sensorid':row[1],'userid':row[2],'checktime':row[3],'verifycode':row[4],'checktype':row[5]})
with open('names.json', 'w') as f:
    json.dump(data,f,indent=4)
```
